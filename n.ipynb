```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ticker = 'NFLX'
url = '' + ticker
stock_data = pd.read_csv("./NFLX.csv")

stock_data
```

```python
d = stock_data
```

```python
change = np.array([])
percent_change = np.array([])

for i in range(len(d)-1):
    today = d.iloc[i].loc['Close']
    yesterday = d.iloc[i+1].loc['Close']
    total_change = today-yesterday
    change = np.append(change, [total_change])
    total_percent_change = str(np.round(change/today*100,2)) + '%'
    percent_change = np.append(percent_change,[total_percent_change])
    
change = np.append(change, ['No previous'])
percent_change= np.append(percent_change, ['No previous day'])
```

```python
change
```

```python
d['Yesterday to Today'] = change
d['Percent Change'] = percent_change
d
```

```python
data = pd.DataFrame(d)
data.to_csv('data.csv')
```

```python
moving_average = np.array([])
dates = np.array([])

for i in range(len(d)-30):
    average = d[i:30+i]['Close'].mean()
    moving_average = np.append(moving_average, [average])
    dates = np.append(dates, [d.iloc[i].loc['Date']])
moving_average
```

```python
m = pd.DataFrame({'30 day moving average': moving_average, 'Date': dates,})
m
```
